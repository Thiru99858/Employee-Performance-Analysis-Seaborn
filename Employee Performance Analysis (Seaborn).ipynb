```python
#Import libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#Import data from local to server and creating data frame
df=pd.read_csv(r"C:\Users\thiru\Downloads\Assessments_thiru\Employee Performance Analysis_02\employee_data.csv.txt")
```

```python
#Checking null values and data types and top 5 data
print(df.head())
df.isnull().sum()
df.dtypes
```

```python
#Histogram + KDE plot

#Creating plot width & height
plt.figure(figsize=(8,4)) 

#Giving plot theme and style
sns.set_style('darkgrid')
sns.set_context('talk')

#Creating Histoplot + KDE plot
sns.histplot(data=df,x='Age',kde=True,color='#5dade2',alpha=0.8,bins=6)

#Creating font dictionary 
f1={'family':'sans-serif','size':'14','color':'#2c3e50','weight':'bold'}
f2={'family':'sans-serif','size':'11','color':'#34495e','weight':'medium'}

#Applying font dictionary and giving names to x,y labels and title
plt.title('Employee Age Distribution',fontdict=f1)
plt.xlabel('Age',fontdict=f2);
plt.ylabel('No.of Employee Count',fontdict=f2);
```

**Insight:** The age distribution shows that most employees fall between 25 and 35 years, indicating a predominantly young and mid-career workforce.

```python
#violinplot

#Creating plot width & height
plt.figure(figsize=(8,4))

#Giving plot theme and style
sns.set_style('darkgrid')
sns.set_context('talk')

#Creating violin plot
sns.violinplot(data=df,x='Department',y='Salary',palette='Set1',hue='Department',legend=False);

#Creating font dictionary 
f3={'family':'sans-serif','size':14,'color':'#2c3e50','weight':'bold'}
f4={'family':'sans-serif','size':11,'color':'#34495e','weight':'medium'}

#Applying font dictionary and giving names to x,y labels and title
plt.title('Salary Distribution by Department',fontdict=f3)
plt.xlabel('Department',fontdict=f4)
plt.ylabel('Salary',fontdict=f4);
```

**Insight:** HR and Finance departments show higher median salaries with wider salary distributions, while IT has comparatively lower salary levels. Sales exhibits a broad salary range, indicating high variation within the department.

```python
#Scatter plot

#Creating plot width & height
plt.figure(figsize=(8,4))

#Giving plot theme and style
sns.set_style('darkgrid')
sns.set_context('paper')

#Creating Scatter plot
sns.scatterplot(data=df,x='Experience',y='Salary',hue='Department',palette='Set1',edgecolor='k',s=50,style='Department')

#Creating font dictionary 
f5={'family':'sans-serif','size':14,'color':'#2c3e50','weight':'bold'}
f6={'family':'sans-serif','size':11,'color':'#34495e','weight':'medium'}

#Applying font dictionary and giving names to x,y labels and title
plt.title('Experience vs Salary Relationship',fontdict=f5)
plt.xlabel('Experience',fontdict=f6)
plt.ylabel('Salary',fontdict=f6);
```

**Insight:** Salary increases steadily with experience across all departments. HR and Finance tend to offer higher salaries at higher experience levels, while Sales shows moderate variation in mid-to-senior experience ranges.

```python
#Count Plot

#Creating plot width & height
plt.figure(figsize=(8,4))

#Giving plot theme and style
sns.set_style('darkgrid')
sns.set_context('notebook')

#Creating Scatter plot
sns.countplot(data=df,x='Performance_Rating',hue='Department');

#Creating font dictionary 
f7={'family':'sans-serif','size':'14','weight':'bold','color':'#2c3e50'}
f8={'family':'sans-serif','size':'11','weight':'medium','color':'#34495e'}

#Applying font dictionary and giving names to x,y labels and title
plt.title('Performance Rating Distribution',fontdict=f7)
plt.xlabel('Performance_Rating',fontdict=f8)
plt.ylabel('Employee Count',fontdict=f8);
```

**Insight:** Most employees receive performance ratings between 3 and 5, indicating generally strong performance levels across departments. HR and Finance show a higher concentration of top ratings, while low ratings are relatively rare.

```python
#Heat Map

#Creating plot width & height
plt.figure(figsize=(8,4)) 

#Aggregation function we are using for grouping and mean values for creating heat map
avg_sal=df.groupby('Performance_Rating')['Salary'].mean().reset_index()

#Here, We are creating index_column for understanding heat map like it is a row values (y-axis type)
Heat=avg_sal.set_index('Performance_Rating')

#Creating heat map
sns.set_context('notebook')
sns.set_style('ticks')
sns.heatmap(data=Heat);

#Creating font dictionary 
f9={'family':'sans-serif','size':14,'weight':'bold','color':'#2c3e50'}
f10={'family':'sans-serif','size':11,'weight':'medium','color':'#34495e'}

#Applying font dictionary and giving names to x,y labels and title
plt.title('Average Salary by Performance Rating',fontdict=f9)
plt.ylabel('Performance_Rating',fontdict=f10);
```

**Insight:** The heatmap shows a clear increase in average salary with higher performance ratings, indicating that employee compensation is strongly linked to performance.
